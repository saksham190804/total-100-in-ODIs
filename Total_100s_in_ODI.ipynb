```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
```

```python
df['Player (Teams Played)'] = pd.factorize(df['Player (Teams Played)'])[0]
```

```python
df.head()
```

```python
df['High Score (* = Not Out)'] = df['High Score (* = Not Out)'].str.replace('*', '').astype(int)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python

# Dropping 'Span' and '100s' columns to create the feature matrix X
X = df.drop(['Span', '100s'], axis=1)

# Assigning the '100s' column values to y
y = df['100s'].values
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# Random seed for reproducibility
mp.random.seed(42)

# Split into train & test set
X_train, X_test, y_train, y_test = train_test_split(X, # independent variables
                                                    y, # dependent variable
                                                    test_size = 0.2) # percentage of data to use for test set
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
# Initialize and fit the Lasso regression model
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, y_train)

# Predict on the test set
y_pred = lasso.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("score:", lasso.score(X_test, y_test))
```

```python
from sklearn.linear_model import ElasticNet
# Initialize and fit the Elastic Net regression model
elastic_net = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)
elastic_net.fit(X_train, y_train)

# Predict on the test set
y_pred = elastic_net.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.linear_model import Ridge
# Initialize and fit the Ridge regression model
ridge = Ridge(alpha=1.0, random_state=42)
ridge.fit(X_train, y_train)

# Predict on the test set
y_pred = ridge.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.svm import SVR
# Initialize and fit the SVR model with linear kernel
svr_linear = SVR(kernel='linear', C=1.0, epsilon=0.1)
svr_linear.fit(X_train, y_train)

# Predict on the test set
y_pred = svr_linear.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Initialize and fit Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)

# Predict on the test set
y_pred = rf_regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# Initialize and fit SVR with RBF kernel
svr_rbf = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr_rbf.fit(X_train, y_train)

# Predict on the test set
y_pred = svr_rbf.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
